```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
# California_House_Dataset
```

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
data
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df
```

```python
df['target'] = data.target
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize =(4,2))
for i in df.columns:
    if df[i].dtype!='object':
        sns.boxplot(df[i])
        plt.title('checking outlabyer')
        plt.xlabel(i)
        plt.show()
```

```python
col = ['Population', 'AveOccup']
for i in col:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    IQR = q3-q1
    ul = q3 + 1.5 * IQR
    ll = q1 - 1.5 * IQR
    
    df = df[(df[i] >= ll) & (df[i] <= ul)]
```

```python
plt.figure(figsize =(4,2))
for i in df.columns:
    if df[i].dtype!='object':
        sns.boxplot(df[i])
        plt.title('checking outlabyer')
        plt.xlabel(i)
        plt.show()
```

```python
x = df.drop('target', axis = 1)
```

```python
x.shape
```

```python
y = df['target']
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 30)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
std_x_train = sc.fit_transform(X_train)
std_x_test = sc.transform(x_test)
```

```python
model = LinearRegression()
model.fit(std_x_train, y_train)
```

```python
y_pre = model.predict(std_x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
r2_score(y_test, y_pre)
```

```python
mean_absolute_error(y_test, y_pre)
```

```python
mean_squared_error(y_test, y_pre)
```
